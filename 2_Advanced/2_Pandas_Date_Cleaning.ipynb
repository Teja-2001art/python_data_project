```python
#Importing Libraries
[1]
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt
#Loading Data
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset ['train'].to_pandas()
#Data Cleanup
df ['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
```

```python
df.loc[:10, 'salary_year_avg' : 'salary_hour_avg']
```

```python
median_salary_year = df['salary_year_avg'].median()
```

```python
median_salary_hour = df['salary_hour_avg'].median()
```

```python
df_filled = df
```

```python
df_filled['salary_year_avg'] = df_filled['salary_year_avg'].fillna(median_salary_year)
df_filled['salary_hour_avg'] = df_filled['salary_hour_avg'].fillna(median_salary_hour)
```

```python
df_filled['salary_year_avg']
```

```python
df_filled['salary_hour_avg']
```

```python
df_filled.loc[:10, 'salary_year_avg':'salary_hour_avg']
```

```python
df_unique = df_filled
df_unique = df_unique.drop_duplicates()
```

```python
print('Lenth of the original df is :', len(df_filled))
print('Length of the drop duplicates df is :', len(df_unique))
print('Length of the removed rows is ',len(df_filled)-len(df_unique) )
```

```python
df_unique_latest = df_unique.drop_duplicates(subset = ['job_title', 'company_name'])
```

```python
print('Lenth of the original df is :', len(df_filled))
print('Length of the drop duplicates df is :', len(df_unique_latest))
print('Length of the removed rows is ',len(df_filled)-len(df_unique_latest) )
```
